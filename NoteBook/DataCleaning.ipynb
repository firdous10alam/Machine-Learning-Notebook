# Feature Engineering and Extraction

```python
df_cleaned['bmi_category'] = pd.cut(
    df_cleaned['bmi'],
    bins = [0,18.5,24.9,29.9,float('inf')],
    labels = ['underweight','normal','overweight','obese']
)
```

```python
df_cleaned = pd.get_dummies(df_cleaned, columns=['bmi_category'])
```

```python
df_cleaned = df_cleaned.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
cols = ['age', 'bmi', 'children']
```

```python
df_cleaned[cols] = scaler.fit_transform(df_cleaned[cols])
```

```python
from scipy.stats import pearsonr
selected_features = ['age', 'is_female', 'bmi', 'children', 'is_smoker',
       'region_northeast', 'region_northwest', 'region_southeast',
       'region_southwest', 'bmi_category_underweight', 'bmi_category_normal',
       'bmi_category_overweight', 'bmi_category_obese']

correlations = {}

for feature in selected_features:
    correlations[feature] = pearsonr(
        df_cleaned[feature],
        df_cleaned['charges']
    )[0]
```

```python
correlation_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Pearson Coefficient']).sort_values(by='Pearson Coefficient', ascending = False)
```

```python
correlation_df
```

```python
correlation_df[correlation_df['Feature'] == 'bmi_category_underweight']['Pearson Coefficient'].values[0]*100
```

```python
from scipy.stats import chi2_contingency

df_cleaned['charges_bins'] = pd.qcut(df_cleaned['charges'], q=4, labels=False)
```

```python
df_cleaned['charges_bins'].value_counts()
```

```python
df_cleaned[['charges', 'charges_bins']]
```

```python
pd.crosstab(df_cleaned['charges'], df_cleaned['charges_bins'])   # Learn it Later
```

```python
df_cleaned.columns
```

```python
col_features = ['is_female', 'is_smoker',
       'region_northeast', 'region_northwest', 'region_southeast',
       'region_southwest', 'bmi_category_underweight', 'bmi_category_normal',
       'bmi_category_overweight', 'bmi_category_obese']
```

```python
col_features
```

```python
chi2_results = {}
alpha = 0.05
for col in col_features:
    contingency = pd.crosstab(df_cleaned[col], df_cleaned['charges_bins'])
    chi2_stat, p_val,_,_ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {'chi2_statistic':chi2_stat, 'p_value':p_val, 'Decision':decision}
```

```python
chi2_df = pd.DataFrame(chi2_results).T     # Here I am getting the Transpose value to analyze well.
chi2_df = chi2_df.sort_values(by='p_value')
```

```python
chi2_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.drop('charges', axis=1)
y = final_df['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
```

```python
print(r2)
```

```python
n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
```
